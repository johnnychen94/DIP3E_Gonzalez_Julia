```python
def hist_eq(img):
    hist = np.zeros(256)
    
    for i in range(0,length):
        for j in range(0,weith):
            temp = img.item(i,j)
            hist[temp] += 1

    pr = np.zeros(256)
    pr = hist/(length*weith)

    s = np.zeros(256)
    temp = 0
    for i in range(0,256):
        temp = temp + pr[i]
        s[i] = round(255*temp)
   
    img1 = np.zeros((length,weith))
        
    for i in range(0,length):
        for j in range(0,weith):
            temp = img[i,j]
            img1[i,j] = s[temp]
    return img1
```

```python
img = mpimg.imread('Fig0320(1)(top_left).tif')
img1 = hist_eq(img)
plt.figure()
plt.subplot(1,2,1)
plt.imshow(img,'gray')
plt.subplot(1,2,2)
plt.imshow(img1,'gray')
plt.figure()
plt.subplot(2,1,1)
n, bins, patches = plt.hist(img.reshape(length*weith), bins=256,edgecolor='None',facecolor='red')
plt.subplot(2,1,2)
n, bins, patches = plt.hist(img1.reshape(length*weith), bins=256,edgecolor='None',facecolor='red')
```
